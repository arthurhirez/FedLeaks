```python
import pandas as pd
import pickle
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open('results/exp_latente_df_graeme.pkl', 'rb') as file:
    latent_dfs = pickle.load(file)
```

```python
for case in latent_dfs.values():
    for epoch in case.values():
        date_cols = epoch['latent_space'].iloc[:, :4].reset_index(drop=True)
        for space in epoch.keys():
            if ('pca' in space) or ('umap' in space):
                epoch[space] = pd.concat([date_cols, epoch[space]], axis = 1)
```

```python
import pandas as pd
import altair as alt
alt.data_transformers.enable("vegafusion")

# Assuming df_latent is already your DataFrame
df = latent_dfs['Baseline'][8]['pca_raw'].copy()

df['week_start'] = pd.to_datetime(df['week'].str.split('/').str[0])
df['week_index'] = df['week_start'].rank(method='dense').astype(int) - 1

# Convert timestamp if needed
df['timestamp'] = pd.to_datetime(df['timestamp'])

week_slider = alt.binding_range(min=df['week_index'].min(), max=df['week_index'].max(), step = 1, name="Release Year")
slider_selection = alt.selection_point(bind=week_slider, fields=['week_index'])

# Base chart using combined data (not grouped per label)
chart = alt.Chart(df).mark_rect().encode(
    x=alt.X('umap_0_scaled:Q', bin=alt.Bin(maxbins=60), title="UMAP 0"),
    y=alt.Y('umap_1_scaled:Q', bin=alt.Bin(maxbins=40), title="UMAP 1"),
    color=alt.Color('label', scale=alt.Scale(scheme='spectral'), title='Density'),
    tooltip=['label:N']
).add_params(
    slider_selection
).transform_filter(
    slider_selection
).properties(
    width=500,
    height=400,
    title="Interactive Latent Space Heatmap (Select Labels)"
)

chart
```
